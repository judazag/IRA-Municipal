## 1. Definición del Problema

Colombia tiene **14,4 millones de personas en inseguridad alimentaria**, pero no existe un instrumento que permita identificar cuáles municipios están al borde del colapso alimentario ni por qué.

Las instituciones responsables (UPRA, alcaldías, organizaciones humanitarias) toman decisiones de focalización con información:
- **Parcial** — solo cubre algunas dimensiones del problema
- **Desactualizada** — no refleja los ciclos agrícolas recientes
- **No integrada** — cada entidad maneja sus propios datos sin cruzarlos

### Pregunta central
> ¿Cuál es el nivel de riesgo alimentario de cada uno de los 1.104 municipios de Colombia y cuál es su cuello de botella específico?


## 2. Objetivos del Proyecto

### Objetivo principal
Construir el **IRA-Municipal**: un índice compuesto que calcule el nivel de riesgo alimentario (Bajo / Medio / Alto / Crítico) para cada uno de los 1.104 municipios de Colombia, cruzando 150 bases de datos abiertas del Estado.

### Objetivos específicos
1. Integrar datos de producción agrícola, acceso logístico, vulnerabilidad climática, pobreza e infraestructura
2. Calcular un índice compuesto con pesos derivados por AHP (Analytic Hierarchy Process)
3. Predecir deterioro del IRA con modelo XGBoost
4. Activar alertas tempranas automáticas ante fenómenos climáticos IDEAM
5. Publicar los resultados en un mapa interactivo y dashboard web


## 3. Criterios de Éxito

| Criterio | Métrica | Meta |
|---|---|---|
| Cobertura municipal | % municipios con IRA calculado | ≥ 95% (1.049/1.104) |
| Consistencia del índice | CR del AHP | < 0.10 |
| Robustez | Variación ranking con pesos ±20% | < 15% municipios cambian nivel |
| Predictivo | F1-score XGBoost | ≥ 0.70 |
| Usabilidad | Tiempo respuesta API | < 2 segundos |


## 4. Dimensiones del IRA

El índice integra **4 dimensiones** complementarias:

### D1 — Producción Agrícola
Mide si el municipio produce suficientes alimentos y si tiene suelos aptos para hacerlo.
- Suficiencia productiva (EVA)
- Diversificación de cultivos (EVA)
- Aptitud de suelos (UPRA — 136 datasets)
- % área dentro de frontera agrícola (UPRA)

### D2 — Acceso y Logística
Mide si los alimentos pueden llegar al municipio y si los agricultores pueden venderlos.
- Flujos de abastecimiento (SIPSA-A DANE)
- Densidad vial rural (OpenStreetMap)
- Cobertura de infraestructura de riego (UPRA)

### D3 — Vulnerabilidad Climática
Mide la exposición del municipio a fenómenos climáticos que afectan la producción.
- Anomalías de precipitación (Estaciones IDEAM)
- Normales climatológicas históricas (IDEAM)
- Índice ONI El Niño/La Niña (NOAA)

### D4 — Condición Socioeconómica
Mide la capacidad de la población para acceder económicamente a los alimentos.
- Índice de Pobreza Multidimensional rural (DANE Censo 2018)
- Necesidades Básicas Insatisfechas rural (DANE Censo 2018)
- Acceso a crédito agropecuario (Finagro)


## 5. Metodología — CRISP-DM

El proyecto adopta **CRISP-DM** (Cross-Industry Standard Process for Data Mining) como marco metodológico.

`
Fase 1: Comprensión del negocio  (este notebook)
Fase 2: Comprensión de los datos (notebook 01_EDA)
Fase 3: Preparación de datos     (notebooks 02_limpieza)
Fase 4: Modelado                 (notebook 04_modelo)
Fase 5: Evaluación               (notebook 04_modelo)
Fase 6: Despliegue               (backend + frontend)
`

CRISP-DM es iterativo: los hallazgos en fases posteriores pueden requerir revisitar fases anteriores.
